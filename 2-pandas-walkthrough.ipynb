# Back to bikes

```python
bikes = pd.read_csv(
    "data/bikes.csv",
    sep=";",
    encoding="latin1",
    parse_dates=["Date"],
    dayfirst=True,
    index_col="Date",
)
bikes["Berri 1"].plot()

```

```python
# double [] is used to return a dataframe (table) instead of a series (column)

berri_bikes = bikes[['Berri 1']].copy()
berri_bikes
```

```python
# index is a datetime!
berri_bikes.index.day
```

```python
# we can access datetime attributes and functions
# for example: weekday, day, month, dayofyear, year, etc.
berri_bikes.index.weekday
```

```python
# create a new column called `weekday` with day names
berri_bikes.loc[:, 'weekday'] = berri_bikes.index.day_name()
berri_bikes
```

```python
weekday_counts = berri_bikes.groupby('weekday').sum()
weekday_counts
```

```python
weekday_counts.plot(kind='barh', figsize=(10, 5))
```

```python
# get weather data of a single month 
```

```python
url_template = "http://climate.weather.gc.ca/climate_data/bulk_data_e.html?stationID=5415&Year={year}&Month={month}&format=csv&timeframe=1&submit=%20Download+Data"

url = url_template.format(month=3, year=2012)
weather_data = pd.read_csv(url, index_col="Date/Time", parse_dates=True,)
weather_data.columns = weather_data.columns.map(lambda x: x.replace("°", ""))
```

```python
weather_data
```

after the first 

```python
bikes.loc[bikes.index.month == 3].plot()
```

```python
# in days between 19 and 26 temperature was very high
# everybody started byclikin

weather_data["Temp (C)"].plot(figsize=(15, 5), title="Temp (C)")
bikes.loc[bikes.index.month == 3].plot(figsize=(15, 5))
```

### aggregate by day (freq = 'D') and take max

```python
weather_data[['Temp (C)']].groupby(pd.Grouper(freq='D')).max()
```

```python
mar_bikes = bikes.loc[bikes.index.month == 3]

# ignore the warning, in this case
mar_bikes['temp'] = weather_data[['Temp (C)']].groupby(pd.Grouper(freq='D')).max()
```

```python
mar_bikes.iloc[:5]
```

```python
corr = mar_bikes.corr()
corr.style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1)
```
